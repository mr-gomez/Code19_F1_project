```python
def iteration(img):
    newimage=img
    m=(img.shape)[0]
    n=(img.shape)[1]
    for i in range(1,m-1):
        for j in range(1,n-1):
            if img[i-1,j,3]*img[i,j-1,3]*img[i,j+1,3]*img[i+1,j,3]==0:
                newimage[i,j,3]=0
    return newimage
```
