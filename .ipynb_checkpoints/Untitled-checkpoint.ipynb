```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.widgets import CheckButtons
from matplotlib.ticker import FuncFormatter

import time

from dfply import *
```

```python
lap_times = pd.read_csv("f1db_csv/lap_times2018.csv")
```

```python
lap_times.head()
```

```python
race = (lap_times >>
        mask(X.raceId==989) >> 
        arrange(X.driverId, X.lap))

drivers = np.unique(race["driverId"])
```

```python
driver_times = (race >> mask(X.driverId==1) >>
                select(X.milliseconds))
```

```python
driver_times['milliseconds'].mean()
```

```python
drivers = np.unique(race["driverId"])
meandrivers=[]
for driver in drivers:
    meandri=[]
    for i in range(989,1010):
        race = (lap_times >>
               mask(X.raceId==i)>>
               arrange(X.driverId))
        driver_times = (race >> mask(X.driverId==driver) >>
                    select(X.milliseconds))
        meandri.append(driver_times['milliseconds'].mean())
        dfmean=pd.DataFrame({'medians':meandri})
    meandrivers.append(dfmean['medians'].mean())
#dfmeandrivers=pd.DataFrame({'medianseason':meandrivers})
#dfmeandrivers
meanseason=pd.DataFrame({'meanseason':meandrivers})
meanseason
```

```python
driverss=pd.DataFrame({'driver':drivers})
```

```python
means_season=driverss.join(meanseason)
```

```python
def driver_name(driverId):
    driver = df_drivers.loc[ df_drivers["driverId"]==driverId, "surname"].values[0]
    
    return driver
```
